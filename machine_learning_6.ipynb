```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('vgsales_6.csv')
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
Oenc.fit(df[['Name']])
df[['Name']] = Oenc.fit_transform(df[['Name']])
df
```

```python
from sklearn.preprocessing import LabelEncoder

Lenc = LabelEncoder()
Lenc.fit(df['Platform'])
df['Platform'] = Lenc.fit_transform(df['Platform'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder

Onhot = OneHotEncoder()
Onhot.fit(df[['Genre']])
Onhot.transform(df[['Genre']]).toarray()
df
```

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()
df = df.replace({'Year' : np.nan}, '2222')
df[['Year']] = Norm.fit_transform(df[['Year']])
df
```

```python
from sklearn.model_selection import train_test_split

Y = df[['Genre', 'Publisher']]
X = df.drop(columns=['Genre', 'Publisher'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

x_train
```
